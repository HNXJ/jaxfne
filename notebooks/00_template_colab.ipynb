## Imports

Import required modules and configure JAX.

```python
import jax
import jax.numpy as jnp

# Verify JAX is available
print(f"JAX version: {jax.__version__}")
print(f"Devices: {jax.devices()}")
```

## Configuration

Set up a minimal jaxfne model. This example uses a small network with Izhikevich neurons.

```python
# Configure a minimal network
cfg = (
    jaxfne.configuration()
    .network(n=10)  # 10 neurons
    .emitter(family="izhikevich", preset="cortical_eig")
    .field(domain="laminar_column", conductivity="proxy")
    .probe(name="tutorial", n_contacts=8)
)

# Construct the model
model = jaxfne.construct(cfg)
print("Model constructed successfully.")
```

## Run Simulation

Simulate 100 milliseconds of neural activity.

```python
# Define simulation parameters
sim = jaxfne.simulation(duration_ms=100.0, dt_ms=0.1)

# Run the simulation
signals = model.simulate(sim)

print(f"Simulation complete.")
print(f"Spike matrix shape: {signals.spikes.shape}")
print(f"Voltage shape: {signals.V_m.shape}")
if hasattr(signals, 'field') and signals.field is not None:
    print(f"LFP proxy shape: {signals.field.lfp_proxy.shape}")
```

## Inspect Output Bundle

Examine the simulation output structure and metadata.

```python
# Check metadata
print("Manifest keys:")
manifest = model.manifest(signals)
for key in list(manifest.keys())[:5]:  # Print first 5 keys
    print(f"  - {key}")

# Check claim status
print(f"\nClaim status:")
print(f"  physical_amplitude_claim_allowed: {manifest.get('physical_amplitude_claim_allowed', False)}")
print(f"  field_claim_level: {manifest.get('field_claim_level', 'proxy_readout_only')}")
print(f"  source_calibration_status: {manifest.get('source_calibration_status', 'uncalibrated')}")
```
